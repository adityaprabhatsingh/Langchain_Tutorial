# **Using Pydantic Models**

```python
from pydantic import BaseModel, Field
# pydantic is libary for data validation and setting management using python type annotation 
class llm_schema(BaseModel):
    setup: str = Field(description="The setup for the joke")
    punchline: str = Field(description="The punchline for the joke")
```

```python
obj=llm_schema(**{"setup":"some setup","punchline": "some punch line"})
```

```python
llm_structured_output = llm_openai.with_structured_output(llm_schema)

result = llm_structured_output.invoke("Tell me a joke")
result.punchline
#return a llm_scheme objeect with thte output of the llm
```

# **Using TypedDict**

```python
from typing import TypedDict 
#exctaly same as pydantic : pydantic->principal(strict) and typedidct(type hint)-> teacher(not much strict)
class llm_schema_td(TypedDict):
    setup: str
    punchline: str
```

Pydantic vs TypedDict
Feature	              TypedDict	                                     Pydantic
Purpose	        Type hints for dictionaries	           Data validation and parsing
Validation	      ❌ No runtime validation                   	✅ Runtime validation
Type Checking	   ✅ Static (mypy, pyright)	                ✅ Static + Runtime
Data Conversion	           ❌ No	                                  ✅ Automatic ("25" → 25)
Performance	           ✅ Faster	                                        ❌ Slightly slower
Dependency	        Standard library	                               External package
Best For	    Internal data structures	                APIs, LLM outputs, user input

```python
llm_structured_typed_dict = llm_openai.with_structured_output(llm_schema_td)

result = llm_structured_typed_dict.invoke("Tell me a joke")
result
```
